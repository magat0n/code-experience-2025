```python
#CONDIÇÕES
#if (se)
    # bloco de código a ser executado se a condição for verdadeira

#elif (senão se)
    # bloco de código a ser executado se a outra condição for verdadeira

#else (senão)
    # bloco de código a ser executado se nenhuma das condições anteriores for verdadeira

nota = int(input("digite sua nota:"))

if nota >= 9:
    print("Aprovado")
elif nota >= 5:
    print("Recuperação")
else:
    print("Reprovado")
```

```python
#.2f :  é usado para formatar números de ponto flutuante (float) com duas casas decimais. Faz parte da sintaxe de formatação de strings.
valor = 10.56789
print(f"O valor formatado é: {valor:.2f}")

#saída
O valor formatado é: 10.57


```

```python
#Geralmente é usado dentro de uma string formatada (f-string) ou com o método format() para exibir números de forma mais legível.

#Como funciona?

# : indica que vamos formatar o valor.
# .2 especifica o número de casas decimais (2 neste caso).
# f indica que estamos formatando um número de ponto flutuante.
```

```python
#Diferenciando Aspas em f-strings:

#Objetivo: Evitar conflitos de sintaxe e melhorar a legibilidade do código.

#Prática: Usar diferentes tipos de aspas para delimitar a f-string e a string literal interna.

#Exemplo: print(f"Olá, {nome}! Você está usando o Colab hoje?")

#Alternativa: Usar aspas triplas para a f-string, permitindo aspas simples e duplas dentro dela sem escape.

#Exemplo: print(f"""Olá, {nome}! Você está usando o "Colab" hoje?""")
```

```python
#Laços de Repetição em Python:

#Objetivo: Executar um bloco de código repetidamente.

#while(enquantov ): Repete enquanto uma condição for verdadeira.
    #Sintaxe: while condição: bloco_de_código

#for (para): Itera sobre uma sequência (lista, tupla, string, range, etc.) Usado para percorrer elementos de uma sequência, e executa um bloco de codigo varias vezes.
    #Sintaxe: for item in sequência: bloco_de_código

#Controle de Fluxo:
#break: Sai do laço prematuramente.
#continue: Pula para a próxima iteração.

#Em resumo, laços while são controlados por condições, enquanto laços for iteram sobre sequências. break e continue permitem controlar o fluxo dentro dos laços.

#quando usar while ao invés do for?

#for quando:
#quando se sabe quantas vezes quer repetir
#está percorrendo listas, strings, etc.

#while quando:
#quando não se sabe quantas vezes
#depende de uma condição que pode mudar
```
