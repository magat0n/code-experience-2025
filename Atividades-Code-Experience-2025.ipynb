Ex. 5
Calculadora simples com verificação de divisão por zero
Peça dois números e uma operação (+, -, *, /). Execute a operação, mas só permita divisão se o segundo número não for zero.

```python
num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))

operacao = input("Digite a operação (+, -, *, /): ")

if operacao == "+":
    resultado = num1 + num2
    print(f"Resultado: {resultado}")
elif operacao == "-":
    resultado = num1 - num2
    print(f"Resultado: {resultado}")
elif operacao == "*":
    resultado = num1 * num2
    print(f"Resultado: {resultado}")
elif operacao == "/":
    if num2 != 0:
        resultado = num1 / num2
        print(f"Resultado: {resultado}")
    else:
        print("Erro: não é possível dividir por zero.")
else:
    print("Operação inválida.")
```

Ex. 6 Verificação de triângulo
Peça ao usuário três valores que representam os lados de um triângulo. Verifique se eles formam um triângulo válido usando a regra:

Um triângulo é válido se a soma de dois lados for sempre maior que o terceiro.

Depois, classifique o triângulo como:

Equilátero: 3 lados iguais

Isósceles: 2 lados iguais

Escaleno: 3 lados diferentes

```python
lado1 = float(input("Digite o primeiro lado do triângulo: "))
lado2 = float(input("Digite o segundo lado do triângulo: "))
lado3 = float(input("Digite o terceiro lado do triângulo: "))

if (lado1 + lado2 > lado3) and (lado1 + lado3 > lado2) and (lado2 + lado3 > lado1):
    print("É um triângulo válido.")

    if lado1 == lado2 == lado3:
        print("Tipo: Equilátero (3 lados iguais)")
    elif lado1 == lado2 or lado1 == lado3 or lado2 == lado3:
        print("Tipo: Isósceles (2 lados iguais)")
    else:
        print("Tipo: Escaleno (3 lados diferentes)")
else:
    print("Os valores não formam um triângulo válido.")
```

Ex.7 Login com tentativa limitada
Crie um sistema de login com:

Usuário: admin

Senha: 1234

O usuário tem 3 tentativas para acertar.

Se errar 3 vezes, mostrar: Acesso bloqueado.

Se acertar antes disso: Bem-vindo, admin!



```python
usuario_correto = "admin"
senha_correta = 1234
tentativas = 3

while tentativas > 0:
    usuario = input("Usuário: ")
    senha = int(input("Senha: "))

    if usuario == usuario_correto and senha == senha_correta:
        print("Bem-vindo, admin!")
        break
    else:
        tentativas -= 1
        print(f"Usuário ou senha incorretos. Tentativas restantes: {tentativas}")

if tentativas == 0:
    print("Acesso bloqueado.")
```

Ex.7 Verificação de número primo
Peça um número e verifique se ele é primo (usando for).

```python
numero = int(input("digite um número: "))

if numero > 1:
    for i in range(2, int(numero/2)+1):
        if (numero % i) == 0:
            print(numero, "não é um número primo.")
            break
    else:
        print(numero, "é um número primo.")

else:
    print(numero, "não é um número primo.")
```

Ex.8 Fatorial com while
Peça um número e calcule seu fatorial usando while.

```python
numero = int(input("Digite um número: "))

fatorial = 1
contador = 1

while contador <= numero:
    fatorial *= contador
    contador += 1

print(f"O fatorial de {numero} é {fatorial}")
```

Ex.9 Jogo de adivinhação com tentativas limitadas
O computador escolhe um número. O usuário tem 5 tentativas para acertar.

```python
numero_secreto = 29
tentativas = 5

print("Você tem 5 tentativas para adivinhar o número secreto de 1 a 50.")

while tentativas > 0:
    palpite = int(input("Digite seu palpite: "))

    if palpite == numero_secreto:
        print("Parabéns! Você acertou o número secreto.")
        break
    elif palpite < numero_secreto:
        print("Seu palpite é menor que o número secreto.")
    else:
        print("Seu palpite é maior que o número secreto.")

    tentativas -= 1
    print(f"Tentativas restantes: {tentativas}")

if tentativas == 0:
    print("Suas tentativas acabaram. O número secreto era", numero_secreto)
```
